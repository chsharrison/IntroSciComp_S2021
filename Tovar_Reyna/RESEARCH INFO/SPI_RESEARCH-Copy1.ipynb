```python
%matplotlib inline

# Creating dataset
x = spi_pm19['speed']
y = spi_pm19["PM 2.5(µg/m³)"]
z = spi_pm19['direction']
 
# Creating figure
fig = plt.figure(figsize = (16, 9))
ax = plt.axes(projection ="3d")
   
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.3,
        alpha = 0.2)
 
 
# Creating color map
my_cmap = plt.get_cmap('hsv')
 
# Creating plot
sctt = ax.scatter3D(x, y, z,
                    alpha =0.9,
                    c = (x + y + z),
                    cmap = my_cmap,
                    marker ='o', s = 100)
 
plt.title("South Padre Island March 2019")
ax.set_xlabel('Speed', fontweight ='bold')
ax.set_ylabel('PM 2.5(µg/m³)', fontweight ='bold')
ax.set_zlabel('Direction', fontweight ='bold')
fig.colorbar(sctt, ax = ax, shrink = 0.5, aspect = 5)
plt.savefig('South Padre Island March 2019')



# show plot
plt.show()

# ax.legend(loc="best");
```

```python
def plot_timeseries(axes, x, y, color, xlabel, ylabel):

  
  axes.plot(x, y, color=color)
  axes.set_xlabel(xlabel)

  
  axes.set_ylabel(ylabel, color=color)

  
  axes.tick_params('y', colors=color)
```

```python
 fig, ax = plt.subplots()


plot_timeseries(ax, spi_pm19.index, spi_pm19["PM 2.5(µg/m³)"], "blue", "March 2019", "PM 2.5 (µg/m^3)")

# Create a twin Axes object that shares the x-axis
ax2 = ax.twinx();

# Plot the relative data in color
plot_timeseries(ax2, spi_pm19.index,spi_pm19["speed"], "green", "March 2019", "Wind Speed (mph)")
ax.set_title("South Padre Island PM 2.5 & Wind Speed for March 2019.")

plt.savefig('spi_wind2019')
```
