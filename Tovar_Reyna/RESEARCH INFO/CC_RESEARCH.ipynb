```python
def plot_timeseries(axes, x, y, color, xlabel, ylabel):

  
  axes.plot(x, y, color=color)
  axes.set_xlabel(xlabel)

  
  axes.set_ylabel(ylabel, color=color)

  
  axes.tick_params('y', colors=color)
```

```python
 fig, ax = plt.subplots()


plot_timeseries(ax, cc_pm19.index, cc_pm19["PM 2.5(µg/m³)"], "black", "March 2019", "PM 2.5 (µg/m^3)")

# Create a twin Axes object that shares the x-axis
ax2 = ax.twinx();

# Plot the relative temperature data in red
plot_timeseries(ax2, cc_pm19.index,cc_pm19["speed"], "green", "March 2019", "Wind Speed (mph)")
ax.set_title("Corpus Christi PM 2.5 & Wind Speed for March 2019.")
```
