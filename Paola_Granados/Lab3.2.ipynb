 E.3 : Contour Demo

A plot such as contour demo would be very useful for plotting contour maps showcasing elevation as the z value, or even concentrations of a contaminant from a point source

```python
#importing necessary additions
import matplotlib
import numpy as np
import matplotlib.cm as cm 
import matplotlib.pyplot as plt
```

```python
delta = 0.025
x = np.arange(-3.0,3.0,delta)
y = np.arange(-2.0,2.0,delta)
X,Y = np.meshgrid(x,y)
Z1 = np.exp(-X**2 - Y**2)
Z2= np.exp(-(X-1)**2-(Y-1)**2)
Z = (Z1-Z2)*2
```

```python
#EXAMPLE 
fig,ax = plt.subplots() #this opens the plot
CS = ax.contour(X,Y,Z) #adding the XYZ values as a contour
ax.clabel(CS,inline=True, fontsize = 10)#adds values of contour circles
ax.set_title("Simplest default with labels")
```

E.4: 3D Surface (colormap)

I find the 3-D surface plots to be very similar to the contour maps, except that they 3 dimensionally show information. These plots can also be used to see where there is a higher concentration of a polutant over a region, and try to find distribution patterns of the contaminants

```python
#importing packages
import matplotlib.pyplot as plt
from matplotlib import cm 
from matplotlib.ticker import LinearLocator
import numpy as np
```

```python
fig, ax = plt.subplots(subplot_kw={"projection":"3d"}) #opens the 3d plot

#make data
X = np.arange(-5,5,0.25)
Y = np.arange(-5,5,0.25)
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

#plot the surface
surf = ax.plot_surface(X,Y,Z,cmap=cm.coolwarm,linewidth = 0, antialiased = False)

#customize the z axis 
ax.set_zlim(-1.01,1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
#a strmethodformatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

#add colorbar
fig.colorbar(surf,shrink=0.5, aspect=5)

plt.show()
```

```python
#Changed the color scheme
#changed from cm.coolwarm to cm.summer

fig, ax = plt.subplots(subplot_kw={"projection":"3d"}) #opens the 3d plot

#plot the surface
surf = ax.plot_surface(X,Y,Z,cmap=cm.summer,linewidth = 0, antialiased = False)

#customize the z axis 
ax.set_zlim(-1.01,1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
#a strmethodformatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

#add colorbar
fig.colorbar(surf,shrink=0.5, aspect=5)

plt.show()
```

E.5 Annotated Heatmap 

The annotated map can be used to visually compare information from categorical axes, such as in this example.
```python
#import necessary packages
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
#Data

vegetables = ["cucumber","tomato","lettuce","asparagus","potato","wheat","barley"]

farmers =["Farmer Joe","Upland Bros","Smith Gardening","Agrifun","Organiculture","BioGoods Ltd.","Cornylee Corp."]

harvest = np.array([[0.8,2.4,2.5,3.9,0.0,4.0,0.0],
                   [2.4,0.0,4.0,1.0,2.7,0.0,0.0],
                   [1.1,2.4,0.8,4.3,1.9,4.4,0.0],
                   [0.6,0.0,0.3,0.0,3.1,0.0,0.0],
                   [0.7,1.7,0.6,2.6,2.2,6.2,0.0],
                    [1.3,1.2,0.0,0.0,0.0,3.2,5.1],
                    [0.1,2.0,0.0,1.4,0.0,1.9,6.3]])
```

E.6: 3D Scatterplot

3-D scatterplots can be used to find correlations between 3 different factors. These factors would have to be numerical or integers. In a scatterplot like this example, spatial information wouldnt be communicated well, unless its manipulated like the surface map. 

```python
#import necessary packages 
import matplotlib.pyplot as plt
import numpy as np
```

```python
#fixing random state 
np.random.seed(19680801)

def randrange(n,vmin,vmax):
    return (vmax-vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

n = 100

for m,zlow,zhigh in [('o',-50,-25),('^',-30,-5)]:
    xs=randrange(n,23,32)
    ys=randrange(n,0,100)
    zs=randrange(n,zlow,zhigh)
    ax.scatter(xs,ys,zs,marker=m)

ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.set_zlabel('Z label')

plt.show()
```

E.7: Countourf and log scale 

The contour plot can be used to plot spatial information. It could be used to show temperatures over an area, or 

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import ma
from matplotlib import ticker, cm
```

```python
N =100
x = np.linspace(-3.0,3.0,N)
y = np.linspace(-2.0,2.0,N)
```

```python
#EXAMPLE
X,Y=np.meshgrid(x,y)
Z1 = np.exp(-X**2 - Y**2)
Z2= np.exp(-(X*10)**2-(Y*10)**2)
z = Z1 + 50*Z2

#put in some negative values
z[:5,:5]=-1

#eliminate warning 
z=ma.masked_where(z<=0,z)

#automatic selection of levels works, setting the log locator
fig,ax = plt.subplots()
cs = ax.contourf(X,Y,z,locator=ticker.LogLocator(),cmap=cm.PuBu_r)

cbar = fig.colorbar(cs)
plt.show()
```

```python
#flip x and y axes
#flipping x and y also changes the axis labels
X,Y=np.meshgrid(x,y)
Z1 = np.exp(-X**2 - Y**2)
Z2= np.exp(-(X*10)**2-(Y*10)**2)
z = Z1 + 50*Z2

#put in some negative values
z[:5,:5]=-1

#eliminate warning 
z=ma.masked_where(z<=0,z)

#automatic selection of levels works, setting the log locator
fig,ax = plt.subplots()
cs = ax.contourf(Y,X,z,locator=ticker.LogLocator(),cmap=cm.PuBu_r)

cbar = fig.colorbar(cs)
plt.show()
```

E.8 Spectrum Representations

Spectrum representations, as the name implies, can be used to show different properties of waves. 

```python
#data
np.random.seed(0)

dt = 0.01
Fs = 1/dt
t = np.arange(0,10,dt)

#generate nnoise
nse = np.random.randn(len(t))
r = np.exp(-t/0.05)
cnse = np.convolve(nse,r)*dt
cnse=cnse[:len(t)]


s = 0.1 * np.sin(4*np.pi*t)+cnse #the signal
```

```python
#EXAMPLE

fig,axs = plt.subplots(nrows=3,ncols=2,figsize=(7,7))

#plot time signal 
axs[0,0].set_title("signal")
axs[0,0].plot(t, s, color='C0')
axs[0,0].set_xlabel("time")
axs[0,0].set_ylabel("amplitude")

#plot different spectrum types
axs[1,0].set_title("Magnitude Spectrum")
axs[1,0].magnitude_spectrum(s,Fs=Fs,color='C1')

axs[1,1].set_title("Log. Magnitude Spectrum")
axs[1,1].magnitude_spectrum(s,Fs=Fs,scale='dB',color='C1')

axs[2,0].set_title("Phase Spectrum ")
axs[2,0].phase_spectrum(s,Fs=Fs,color='C2')

axs[2,1].set_title("Angle Spectrum")
axs[2,1].angle_spectrum(s,Fs=Fs,color='C2')

axs[0, 1].remove()  # don't display empty ax

fig.tight_layout()
plt.show()
```
