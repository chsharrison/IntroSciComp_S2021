```python
#split into 2 groups based on location 
air_alton = air_data[air_data["Location"] == "Alton, Texas"]
air_SantaAna = air_data[air_data["Location"] != "Alton, Texas"]
```

```python
#make a function that plots concentrations 

#I dont know why alton wont be plotted 
#It will only take the last site inputted

def scatter_airdate(site1,site2,targetgas):
    fig, ax = plt.subplots()
    #plt.figure()
    x1=site1["Date"]
    y3=site2[targetgas]
    a1=site2["Date"]
    b3=site2[targetgas]
    plt.subplot()
    plt.scatter(x1,y3)
    plt.scatter(a1,b3)
    plt.xticks(rotation=45)
    plt.legend()
    plt.title(targetgas)
```

```python
scatter_airdate(air_alton,air_SantaAna,"DMS (MS)")
```

```python
scatter_airdate(air_alton,air_SantaAna,"Benzene (E/B/MS)")
```

```python
def scatter_air_pd (site1,site2,targetgas):
    #fig, ax = plt.subplots()
    site1.plot.scatter(x = "Date", y = targetgas)
    site2.plot.scatter(x = "Date", y = targetgas)
```

```python
scatter_air_pd(air_alton,air_SantaAna,"DMS (MS)")
```
