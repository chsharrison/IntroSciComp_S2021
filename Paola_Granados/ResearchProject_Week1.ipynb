```python
#plotting annual avg temp vs year will give us a general idea 
plt.plot(mcmill_month["Year"],mcmill_month["Annual"], label = "McAllen Millet Intl Airport")
```

```python
#linear fit for annual temperatures 
from scipy import stats 
annualtemptrend = stats.linregress(mcmill_month["Year"],mcmill_month["Annual"])
```

```python
annualtemptrend.intercept
```

```python
plt.plot(mcmill_month["Year"],mcmill_month["Annual"], label = "McAllen Millet Intl Airport")
plt.plot(mcmill_month["Year"],annualtemptrend.slope*mcmill_month["Year"]+annualtemptrend.intercept)
plt.xlabel("Year")
plt.ylabel("Annual Average Temperature ($\\degree$F)")
plt.legend()
#plt.set_title("Change in local temperatures over time ")
```

```python
#pseudocode
#go through every row
#for each row obtain the column name and value 
    #array[r,colname,colvalue]
#in a new array add year+month+avgtemp

array = []
years = []
motnhs =[]
for r in mcmill_month["Year"]:
    for c in mcmill_month:
        array = array + [r]+[c]
        years =years + [r] #creates an array with the year for each datapoint
        motnhs=motnhs + [c] #creates an array with month for each datapoint
        #array = np.append()
```
