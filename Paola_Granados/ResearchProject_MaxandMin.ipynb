compare rate of change of Tmin and Tmax 

```python
from scipy import stats
```

```python
#linear regression for tmax
tmaxtemptrend = stats.linregress(McMill_tmax["Year"],McMill_tmax["Annual"])
```

```python
tmaxtemptrend.slope
```

```python
#linear regression for tmin
tmintemptrend = stats.linregress(McMill_tmin["Year"],McMill_tmin["Annual"])
```

```python
tmintemptrend.slope
```

The tmins increase much slower than tmax 
I think tmax is similar to avg 

```python
#make equations for trends
annualtmax = (tmaxtemptrend.slope*McMill_tmax["Year"])+tmaxtemptrend.intercept
annualtmin = (tmintemptrend.slope*McMill_tmin["Year"])+tmintemptrend.intercept
```

```python
#plot the linear trends 
plt.plot(McMill_tmax["Year"],McMill_tmax["Annual"], linestyle =":", c = "r")
plt.plot(McMill_tmax["Year"],annualtmax,c = "r" )

```

```python
plt.plot(McMill_tmin["Year"],McMill_tmin["Annual"],linestyle = ":",c = "b")
plt.plot(McMill_tmin["Year"], annualtmin, c="b")
```

```python
plt.plot(McMill_tmax["Year"],McMill_tmax["Annual"], linestyle =":", c = "r")
plt.plot(McMill_tmax["Year"],annualtmax,c = "r" )
plt.plot(McMill_tmin["Year"],McMill_tmin["Annual"],linestyle = ":",c = "b")
plt.plot(McMill_tmin["Year"], annualtmin, c="b")
```

```python
#try a split in y-axis so we can compare the two trends 
f, (ax,ax2) = plt.subplots(2,1,sharex = True)

#plot the same data on both axes

#plot1
ax.plot(McMill_tmax["Year"],McMill_tmax["Annual"], linestyle =":", c = "r")
ax.plot(McMill_tmax["Year"],annualtmax,c = "r", label = "Annual TMAX" )
ax.plot(McMill_tmin["Year"],McMill_tmin["Annual"],linestyle = ":",c = "b")
ax.plot(McMill_tmin["Year"], annualtmin, c="b", label = "Annual TMIN")


#plot2
ax2.plot(McMill_tmax["Year"],McMill_tmax["Annual"], linestyle =":", c = "r")
ax2.plot(McMill_tmax["Year"],annualtmax,c = "r", label = "Annual TMAX" )
ax2.plot(McMill_tmin["Year"],McMill_tmin["Annual"],linestyle = ":",c = "b")
ax2.plot(McMill_tmin["Year"], annualtmin, c="b", label = "Annual TMIN")
ax2.legend()


#set the limits to different portions of data 
ax.set_ylim(98,112)#tmax
ax2.set_ylim(18,40)#tmin

#erase the spines between ax and ax2
ax.spines["bottom"].set_visible(False)
ax2.spines["top"].set_visible(False)
ax.xaxis.tick_top()
ax.tick_params(labeltop = False)
ax2.xaxis.tick_bottom()

#make labels
plt.xlabel("Year")
#plt.ylabel("Temperature ($\degree$F) ")

#making shared y label 
f.add_subplot(111, frame_on = False)
plt.tick_params(labelcolor = "none", bottom = False, left = False)
plt.ylabel("Temperature ($\degree$F)")

plt.show()

```

what about detrending the data 

```python
#detrending tmax and tmin temps 
tmaxdetrend = McMill_tmax["Annual"]-annualtmax
tmindetrend = McMill_tmin["Annual"]-annualtmin
```

```python
#Set plotting formats
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = [13.0, 6.0]
mpl.rcParams['font.size'] = 16
mpl.rcParams['legend.fontsize'] = 'medium'
mpl.rcParams['figure.titlesize'] = 'medium'
mpl.rcParams['lines.linewidth'] = 2.0
```

```python
#plot detrended data 
plt.plot(McMill_tmax["Year"],tmaxdetrend,c="r", label = "Detrended Monthly TMAX")
plt.plot(McMill_tmin["Year"],tmindetrend,c="b", label = "Detrended Monthly TMIN")
plt.legend()
```
