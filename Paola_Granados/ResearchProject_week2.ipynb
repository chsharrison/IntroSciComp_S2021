```python
calday_tmax = pd.read_csv("C:/Users/Pgran/Documents/ENVRClasses/MARS4300/calday_tmax.csv")
```

```python
calday_tmax
```

```python
#fix into a dataset we can plot 

calday = []
months = []

for r in calday_tmax["Day"]:
    for c in calday_tmax:
        calday = calday + [r]
        months = months + [c]
```

```python
#make days and months into dataframes so we can join them 
caldaydf = pd.DataFrame(calday)
monthsdf = pd.DataFrame(months)
```

```python
caldaydf
```

```python
#get the datapoints - the year in which tmax ocurred 

calday_tmax_t = calday_tmax.transpose()#transpose the dataset

calday_tmaxdata = []
for c in calday_tmax_t:
    calday_tmax_c = calday_tmax_t[c].values.tolist()
    calday_tmaxdata = calday_tmaxdata + calday_tmax_c
    
calday_tmaxdata
```

```python
calday_tmaxdatadf = pd.DataFrame(calday_tmaxdata)
```

```python
#Join days and month
Caldata_TMAX = caldaydf.join(monthsdf, on = None, how = "left", lsuffix = "day", rsuffix = "month")
Caldata_TMAX
```

```python
#join CALDATA TMAX with data(Years)
Caldata_TMAX = Caldata_TMAX.join(calday_tmaxdatadf, on = None, how = "left", lsuffix = "left",rsuffix = "right")
Caldata_TMAX
```

```python
#rename columns 
Caldata_TMAX = Caldata_TMAX.rename(columns = {"0day":"Day","0month":"Month",0:"Year_of_TMAX"})
```

```python
Caldata_TMAX
```

```python
#remove rows where day = tmax 

#select rows where day does not equal tmax 
Caldata_TMAX = Caldata_TMAX[Caldata_TMAX["Day"]!=Caldata_TMAX["Year_of_TMAX"]]
Caldata_TMAX
```

```python
#Need to find the number of TMAX values each yeas has 
tmax_per_yr = Caldata_TMAX.groupby("Year_of_TMAX").count()
tmax_per_yr = tmax_per_yr.drop(0)
tmax_per_yr
```

```python
calday_tmin = pd.read_csv("C:/Users/Pgran/Documents/ENVRClasses/MARS4300/calday_tmin.csv")
```

```python
calday_tmin
```

```python
#get the datapoints - the year in which tmax ocurred 

calday_tmin_t = calday_tmin.transpose()#transpose the dataset

calday_tmindata = []
for c in calday_tmin_t:
    calday_tmin_c = calday_tmin_t[c].values.tolist()
    calday_tmindata = calday_tmindata + calday_tmin_c
    
calday_tmindata
```

```python
#make tmin datapoints into a dataframe
calday_tmindatadf = pd.DataFrame(calday_tmindata)
```

```python
#Join Data 
Caldata_TMIN = caldaydf.join(monthsdf, on = None, how = "left", lsuffix = "day", rsuffix = "month")
Caldata_TMIN = Caldata_TMIN.join(calday_tmindatadf, on = None, how = "left", lsuffix = "left",rsuffix = "right")
Caldata_TMIN
```

```python
#rename columns 
Caldata_TMIN = Caldata_TMIN.rename(columns = {"0day":"Day","0month":"Month",0:"Year_of_TMIN"})
```

```python
#remove rows where day = tmax 

#select rows where day does not equal tmax 
Caldata_TMIN = Caldata_TMIN[Caldata_TMIN["Day"]!=Caldata_TMIN["Year_of_TMIN"]]
Caldata_TMIN
```

```python
#Need to find the number of TMAX values each yeas has 
#Day and Month values are in this case the number of tmin records within each year
tmin_per_yr = Caldata_TMIN.groupby("Year_of_TMIN").count()
tmin_per_yr = tmin_per_yr.drop(0)
tmin_per_yr
```

```python
tmin_per_yr.plot()
tmax_per_yr.plot()
```

```python
#Plot number of daily records per year 
plt.scatter(tmin_per_yr.index, tmin_per_yr["Day"], c = "b", label = "Record TMIN")
plt.scatter(tmax_per_yr.index, tmax_per_yr["Day"], c = "r", label = "Record TMAX")
plt.xlabel("Year")
plt.ylabel("# of daily records")
plt.legend()
```
