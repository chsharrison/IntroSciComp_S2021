```python
import matplotlib.pyplot as plt
plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt


plt.rcdefaults()
fig, ax = plt.subplots()

people = ('Ram', 'Ashim', 'Shyam', 'Sahil', 'Suman')
y_sec = np.arange(len(people))
km = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

ax.barh(y_sec, km, xerr=error, align='center',
        color='yellow', ecolor='black')
ax.set_yticks(y_sec)
ax.set_yticklabels(people)
ax.invert_yaxis()  
ax.set_xlabel('Kilometer')
ax.set_title('Kilometer a player cover in a football ')

plt.show()
```

```python
import matplotlib.pyplot as plt
country = 'England', 'Brazil', 'Portugal', 'Other'
num = [44, 22, 18, 16]
color = 'yellow', 'green', 'blue', 'red'
esc = (0.5, 0, 1, 1.5)

fig1, ax1 = plt.subplots()
ax1.pie(num, esc, country, color, autopct='%10.1f%%', shadow=True, startangle=90)

ax1.axis('equal')
plt.title('Manchester United Players players from country in Pie Chart')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.pi
start = -2*np.pi
stop = 2*np.pi
step = np.pi/128


x = np.arange(start, stop+step, step)

y = np.sin(x)
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('The Curve of sinX')
plt.show()
```
