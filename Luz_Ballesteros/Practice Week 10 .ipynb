```python
##drop the NaNs
#drop nans in the target columns 
movies1_dropna = movies_sheet1.dropna(subset = ["Reviews by Users","Reviews by Crtiics"])
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(movies1_dropna["Reviews by Users"],movies1_dropna["Reviews by Crtiics"])
```

```python
plt.plot(movies1_dropna["Reviews by Users"], slope*movies1_dropna["Reviews by Users"]+intercept, 'k--') # y = mx+b
plt.scatter(movies1_dropna["Reviews by Users"],movies1_dropna["Reviews by Crtiics"])
plt.xlabel('Reviews by Users')
plt.ylabel('Reviews by Critics')
plt.annotate("y = 0.196x+86.49",(3000,600))
```

```python
np.corrcoef(movies1_dropna["Reviews by Users"],movies1_dropna["Reviews by Crtiics"])
```

```python
## using the OLS to get a nice chart 
X = movies1_dropna["Reviews by Users"]
y = movies1_dropna["Reviews by Crtiics"]
```

```python
import statsmodels.api as sm
```

```python
X=sm.add_constant(X)
```

```python
model = sm.OLS(y, X).fit()
```

```python
predictions = model.predict(X)
```
