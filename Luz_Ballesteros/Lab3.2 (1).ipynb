```python
# let's go over the first example plot in the matplotlib documenation gallery

labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
men_std = [2, 3, 4, 1, 2]
women_std = [3, 5, 2, 3, 3]

width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, men_means, width, yerr=men_std, label='Men')
ax.bar(labels, women_means, width, yerr=women_std, bottom=men_means,
       label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()
```

```python
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi *t)
```

```python
fig, ax = plt.subplots()
ax.plot(t, s)
ax.set(xlabel='time (s)', ylabel='voltage (mV)', title='Simple but not really')
ax.grid()
fig.savefig("first.png")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(t, s, color='r', linewidth=3.0)
ax.set(xlabel='time (s)', ylabel='voltage (mV)', title='Change Color')
ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(t, s, color='y', linewidth=3.0, ls='--')
ax.set(xlabel='time (s)', ylabel='Speed', title='Bee Path Velocity')
ax.grid()
plt.show()
```

```python
# contour plot 
delta = 0.025 
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X,Y = np.meshgrid(x,y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(-X-1)**2-(Y-1)**2)
Z = (Z1-Z2)*2
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X,Y,Z)
ax.clabel(CS,inline=True, fontsize = 20)
ax.set_title("Default Labels")
```

```python
# you can change the font size as well as the title
fig, ax = plt.subplots()
CS = ax.contour(X,Y,Z)
ax.clabel(CS,inline=True, fontsize = 5)
ax.set_title("Contour")
```

```python
# you can also flip the axes by changing the definition of CS
fig, ax = plt.subplots()
CS = ax.contour(Y,Z,X)
ax.clabel(CS,inline=True, fontsize = 10)
ax.set_title("Flip Axis")
```

```python
X = np.arange(-5,5,0.25)
Y = np.arange(-5,5,0.25)
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
```

```python
fig, ax = plt.subplots(subplot_kw={"projection":"3d"})

surf = ax.plot_surface(X,Y,Z,cmap=cm.coolwarm,linewidth = 0, antialiased = False)
```

```python
# add a color bar 
fig, ax = plt.subplots(subplot_kw={"projection":"3d"})

surf = ax.plot_surface(X,Y,Z,cmap=cm.coolwarm,linewidth = 0, antialiased = False)

fig.colorbar(surf,shrink=0.5, aspect=5)

plt.show()
```

```python
# change the color scheme:
fig, ax = plt.subplots(subplot_kw={"projection":"3d"})

surf = ax.plot_surface(X,Y,Z,cmap=cm.summer,linewidth = 0, antialiased = False)
fig.colorbar(surf,shrink=0.5, aspect=5)

plt.show()
```

```python
year = [1950,1960,1970,1980,1990,2000,2010,2018]
population_by_continent = {
    'africa':[228,284,365,477,631,814,1044,1275],
    'americas':[340,425,519,619,727,840,943,1006],
    'asia':[1394,1686,2120,2625,3202,3714,4169,4560]}
```

```python
fig,ax = plt.subplots()
ax.stackplot(year,population_by_continent.values(),
            labels=population_by_continent.keys())
ax.legend(loc='upper left')
ax.set_title('World population')
ax.set_xlabel('Year')
ax.set_ylabel('Number of people(millions)')

plt.show()
```

```python
## you can move the legend by using the ax.legend(loc=)
fig,ax = plt.subplots()
ax.stackplot(year,population_by_continent.values(),
            labels=population_by_continent.keys())
ax.legend(loc='lower right')
ax.set_title('World population')
ax.set_xlabel('Year')
ax.set_ylabel('Number of people(millions)')

plt.show
```
